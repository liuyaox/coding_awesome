### 2.1 Basic Operations on Images

**Goals**

- Access pixel values and modify them

- Access image properties

Note: Almost all the operations in this section is mainly related to Numpy rather than OpenCV.

```python
img = cv2.imread('demo.jpg', flags=1)
print(img.shape)
print(img.size)  # 733*790*3
print(img.dtype) # Very important because many errors in OpenCV-Python code are caused by invalid datatype.
```

```python
# This's normally used for selecting a region of array, say first 5 rows and last 3 columns 
bgr = img[100,100]              #获得坐标处的BGR
img[100, 100] = [255, 255, 255] #修改坐标处的BGR

# For individual pixel access and modification, we recommend methods: array.item/itemset
r = img.item(100, 100, 2)
img.itemset((100, 100, 2), 255)
```

```python
# Split the BGR channels, then merge to RGB model
# cv2.split is a costly operation in terms of time. So do it only if you need it. Otherwise go for Numpy indexing.
b, g, r = cv2.split(img)
# b = img[:, :, 0]
img2 = cv2.merge([r, g, b])
```

```python
# Image ROI: Get certain region of image
region1 = img2[220: 250, 130: 260]
plt.imshow(region1)
plt.show()
img2[400: 430, 400: 530] = region1
plt.imshow(img2)
plt.show()
```

```python
# Making Borders for Images (Padding) - cv2.copyMakeBorder
BLUE = [255, 0, 0]
img1 = cv2.imread('opencv_logo.png')

replicate = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_CONSTANT,value=BLUE)

plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')
plt.show()
```

### 2.2 Arithmetic Operations on Images

**Goals**

- Learn several arithmetic operations on images like addition, subtraction, bitwise operations etc.

- You will learn these functions : cv2.add(), cv2.addWeighted() etc.

```python
# Image Addition - 颜色上的叠加
# OpenCV addition is a saturated operation while Numpy addition is a modulo operation
x = np.uint8([250])
y = np.uint8([10])
print(cv2.add(x, y))  # 250+10 = 260 => 255
print(x + y)          # 250+10 = 260 % 256 = 4
```

```python
# Image Blending - 颜色上的加权叠加 类似水印
# Another image addition, but different weights are given to images, giving a feeling of blending or transparency
# Images are added as per the equation: g(x) = alpha * f0(x) + beta * f1(x) + gamma
img1 = cv2.imread('./images/2.2_1.jpg')
img2 = cv2.imread('./images/2.2_2.jpg')
img3 = cv2.addWeighted(img1, 0.3, img2, 0.7, 0)
plt.subplot(131),plt.imshow(img1, 'gray')
plt.subplot(132),plt.imshow(img2, 'gray')
plt.subplot(133),plt.imshow(img3, 'gray')
plt.show()
```

### 3.4 Smoothing Images

**Goals**

- Blur the images with various low pass filters

- Apply custom-made filters to images (2D convolution)

```python
# 2D Convolution (Image Filtering) - 卷积
# Low-pass  filters(LPF) helps in removing noises, blurring the images etc.
# High-pass filters(HPF) helps in finding edges in the images.
img = cv2.imread('./images/opencv-logo.png')

kernel = np.ones((7,7),np.float32)/49
dst = cv2.filter2D(img ,-1, kernel)   # The function does actually compute correlation, not convolution

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Image Blurring (Image Smoothing) - 4 types of blurring techniques
# Averaging
blur = cv2.blur(img,(5,5))
# Gaussian Blurring
blur = cv2.GaussianBlur(img,(5,5),0)
# Median Blurring
median = cv2.medianBlur(img,5)
# Bilateral Filtering
blur = cv2.bilateralFilter(img,9,75,75)
```

### 3.5 Morphological Transformations

**Goals**

- We will learn different morphological operations like Erosion, Dilation, Opening, Closing etc.

- We will see different functions like : cv2.erode(), cv2.dilate(), cv2.morphologyEx() etc.

```python
img = cv2.imread('./images/j.png', 1)
img = img[:, :, ::-1]
kernel = np.ones((5,5),np.uint8)

erosion  = cv2.erode(img, kernel, iterations=1)
dilation = cv2.dilate(img, kernel, iterations=1)
opening  = cv2.morphologyEx(img, cv2.MORPH_OPEN,     kernel) # =Erosion followed by dilation. Useful in removing noise
closing  = cv2.morphologyEx(img, cv2.MORPH_CLOSE,    kernel) # =Dilation followed by erosion. Useful in closing small holes or small black points
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
tophat   = cv2.morphologyEx(img, cv2.MORPH_TOPHAT,   kernel)
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)

plt.subplot(231), plt.imshow(img), plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(232),plt.imshow(erosion),plt.title('Erosion'), plt.xticks([]), plt.yticks([])
plt.subplot(233),plt.imshow(dilation),plt.title('Dilation'), plt.xticks([]), plt.yticks([])
plt.subplot(234),plt.imshow(gradient),plt.title('Gradient'), plt.xticks([]), plt.yticks([])
plt.subplot(235),plt.imshow(tophat),plt.title('Tophat'), plt.xticks([]), plt.yticks([])
plt.subplot(236),plt.imshow(blackhat),plt.title('Blackhat'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Get Structuring Element.- Just pass the shape and size of the kernel, you get the desired kernel.
print(cv2.getStructuringElement(cv2.MORPH_RECT,    (5,5)))
print(cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5)))
print(cv2.getStructuringElement(cv2.MORPH_CROSS,   (5,5)))
```

### 3.13 Hough Line Transform

**Goals**

- We will understand the concept of Hough Tranform.

- We will see how to use it detect lines in an image.

- We will see following functions: cv2.HoughLines(), cv2.HoughLinesP()

**Theory**

Hough Transform is a popular technique to detect any shape, if you can represent that shape in mathematical form. It can detect the shape even if it is broken or distorted a little bit.

```python
img = cv2.imread('demo.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 400, 500, apertureSize=3)

# Finds lines in a binary image using the standard Hough transform.
img2 = img.copy()
lines = cv2.HoughLines(image=edges, rho=1, theta=np.pi/180, threshold=200)
for rho, theta in lines[0]:
    a, b = np.cos(theta), np.sin(theta)
    x0 = a * rho
    y0 = b * rho
    x1 = int(x0 + 1000 * (-b))
    y1 = int(y0 + 1000 * a)
    x2 = int(x0 - 1000 * (-b))
    y2 = int(y0 - 1000 * a)
    cv2.line(img2, (x1, y1), (x2, y2), (0 ,0, 255), 2)
cv2.imwrite('demo3.jpg', img2)

# Finds line segments in a binary image using the probabilistic Hough transform.
img2 = img.copy()
lines = cv2.HoughLinesP(image=edges, rho=1, theta=np.pi/180, threshold=100, minLineLength=100, maxLineGap=10)
for x1, y1, x2, y2 in lines[0]:
    cv2.line(img2, (x1, y1), (x2, y2), (0, 255, 0), 2)
cv2.imwrite('demo4.jpg', img2)
```
